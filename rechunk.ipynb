```python
import zarr
from rechunker import rechunk
import nd2
from dask.distributed import Client, LocalCluster
cluster = LocalCluster(
    processes=True,       
    threads_per_worker=1, 
    n_workers=16,          
    memory_limit='16GB'    
)
client = Client(cluster)
```

```python
experiment_nd2 = nd2.imread("/data/20251127_SB1_M9-0.05_24hrsOld_1hrAmpicillin_differentConcs.nd2", dask=True, xarray=True)
```

```python
experiment_nd2.dims
```

```python
intermediate = "/data/intermediate.zarr"

target = "/data/target.zarr"

target_chunks = {
    "T" : 1,
    "P" : 1,
    "X" : 2304,
    "Y" : 2304
}

rechunked = rechunk(experiment_nd2, target_chunks=target_chunks, target_store=target,

                    max_mem="100MB",

                    temp_store=intermediate)

rechunked
```

```python
result = rechunked.execute()
```
